# Step 1 - Loading the Required Libraries and Modules

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn

# Import necessary modules
from sklearn.model_selection import train_test_split

# Keras specific
import keras
from keras.models import Sequential
from keras.layers import Dense
```

# Step 2 - Reading the Data and Performing Basic Data Checks

```python
df = pd.read_csv('diabetes.csv')
print(df.shape)
```

```python
df.describe()
```

```python
df.columns
```

# Step 3 - Creating Arrays for the Features and the Response Variable

```python
X = df[df.columns[0:8]]
y = df['Outcome']
```

# Step 4 - Creating the Training and Test Datasets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
print(X_train.shape); print(X_test.shape)
```

# Step 5 - Define, Compile, and Fit the Keras Classification Model

```python
model = Sequential()
model.add(Dense(500, activation='relu', input_dim=8))
model.add(Dense(500, activation='relu'))
model.add(Dense(500, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# build the model
model.fit(X_train, y_train, epochs=100)
```

# Step 6 - Predict on the Test Data and Compute Evaluation Metrics

```python
model.evaluate(X_test, y_test)
```

# Model Building


```python
# Model Building
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Activation
model = Sequential()
# reshape into 1-dimensional data
model.add(Flatten(input_shape=(28, 28)))

# dense layer 1
model.add(Dense(256, activation='relu'))

# dense layer 2
model.add(Dense(128, activation='relu'))

# output layer
model.add(Dense(10, activation='sigmoid'))

# Model summary
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="Adam",metrics=["accuracy"])
model.fit(X_train,y_train,epochs=25,validation_split=0.2)
```

# Model Prediction

```python
results = model.evaluate(X_test,  y_test)
print('test loss, test acc:', results)
```
