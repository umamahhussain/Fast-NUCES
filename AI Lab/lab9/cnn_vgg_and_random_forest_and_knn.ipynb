```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Preprocess the data
x_train = x_train.astype('float32') / 255.0  # Normalize pixel values to be between 0 and 1
x_test = x_test.astype('float32') / 255.0

# Define the CNN model using Sequential API
model = Sequential()

# Add Convolutional layers
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))  # 32 filters of 3x3 with 'relu' activation
model.add(MaxPooling2D((2, 2)))  # Max pooling layer with 2x2 pool size

model.add(Conv2D(64, (3, 3), activation='relu'))  # Another Convolutional layer
model.add(MaxPooling2D((2, 2)))  # Max pooling layer

# Flatten the output before feeding into Dense layers
model.add(Flatten())

# Add Dense layers for classification
model.add(Dense(128, activation='relu'))  # 128 neurons in the dense layer with 'relu' activation
model.add(Dense(10, activation='softmax'))  # Output layer with 10 neurons for 10 classes (using 'softmax' for multi-class classification)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Display model summary
model.summary()

# Train the model
model.fit(x_train, y_train, epochs=1, batch_size=64, validation_split=0.1)

# Evaluate the model on test data
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_accuracy}")
```

```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import numpy as np

# Load CIFAR-10 dataset
(x_train_full, y_train_full), (x_test, y_test) = cifar10.load_data()

# Preprocess the data
x_train_full = x_train_full.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Convert labels to one-hot encoding
num_classes = 10
y_train_full = to_categorical(y_train_full, num_classes)
y_test = to_categorical(y_test, num_classes)

# Split the training data into training and validation sets
x_train, x_val, y_train, y_val = train_test_split(x_train_full, y_train_full, test_size=0.1, random_state=42)

# Load VGG16 model pretrained on ImageNet without the top (fully connected) layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))

# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Create a new model on top of the pretrained base model
model = Sequential()
model.add(base_model)
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Display model summary
model.summary()

# Train the model
history = model.fit(x_train, y_train, epochs=1, batch_size=64, validation_data=(x_val, y_val))

# Evaluate the model on test data
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_accuracy}")
```

```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# Load CIFAR-10 dataset
(x_train_full, y_train_full), (x_test, y_test) = cifar10.load_data()

# Preprocess the data
x_train_full = x_train_full.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Convert labels to one-hot encoding
num_classes = 10
y_train_full = to_categorical(y_train_full, num_classes)
y_test = to_categorical(y_test, num_classes)

# Split the training data into training and validation sets
x_train, x_val, y_train, y_val = train_test_split(x_train_full, y_train_full, test_size=0.1, random_state=42)

# Load VGG16 model pretrained on ImageNet without the top (fully connected) layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))

# Extract features using the VGG16 model
x_train_features = base_model.predict(x_train)
x_val_features = base_model.predict(x_val)
x_test_features = base_model.predict(x_test)

# Flatten the extracted features
x_train_flatten = x_train_features.reshape(x_train_features.shape[0], -1)
x_val_flatten = x_val_features.reshape(x_val_features.shape[0], -1)
x_test_flatten = x_test_features.reshape(x_test_features.shape[0], -1)

# Initialize k-NN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)

# Train the k-NN classifier using the extracted features
knn_classifier.fit(x_train_flatten, np.argmax(y_train, axis=1))

# Evaluate the k-NN classifier on validation data
val_predictions = knn_classifier.predict(x_val_flatten)
val_accuracy = accuracy_score(np.argmax(y_val, axis=1), val_predictions)
print(f"Validation Accuracy: {val_accuracy}")

# Evaluate the k-NN classifier on test data
test_predictions = knn_classifier.predict(x_test_flatten)
test_accuracy = accuracy_score(np.argmax(y_test, axis=1), test_predictions)
print(f"Test Accuracy: {test_accuracy}")
```
